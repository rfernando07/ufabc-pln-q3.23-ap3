#### **INTRODUÇÃO**

A AwsomeAPI é um API gratuita que possui aplicações para cotação de moedas. Uma API de cotações é uma ferramenta essencial para qualquer pessoa, empresa ou aplicativo que deseja acessar informações atualizadas sobre preços de ativos financeiros, como ações, moedas, commodities ou criptomoedas.

Essas APIs são uma fonte valiosa de dados em tempo real ou históricos que permitem aos usuários monitorar e analisar os mercados financeiros de maneira eficaz, sendo importante para a tomada de decisões financeiras, acompanhamento de portifólio e análise de mercado, pois tanto investidores, traders e gestores de portfólio dependem de informações precisas sobre cotações para tomar decisões financeiras. Elas também são podem ser uteis para o desenvolvimento de aplicativos financeiros para integrar informações atualizadas para serem utilizadas em negociação de ações, aplicativos de análise de mercado e serviços de consultoria.

#### **UTILIZAÇÃO BÁSICA**

Para utilizar a API é necessários realizar requisições HTTP do tipo **GET**, que serve para recuperar dados, para os links especificos de cada função disponível, sempre inserindo no mesmo link de requisição os parâmetros necessários para a informação que se deseja extrair. A seguir, vamos demonstrar dois exemplos de utilização, obtendo a cotação de outras moedas e como tratar os dados que vem da requisição.

#####Convertendo Dolar e Euro para Real

Para demonstrar essa utilização básica devemos utiizar o link abaixo, ele retornará a atual cotação de uma moeda especifica para uma moeda de destino e é atualizado a cada 30 segundos.

Link: https://economia.awesomeapi.com.br/json/last/:moedas

No parâmetro moedas ":moedas" que está no link, devemos inserir os códigos das moedas que desejamos realizar as conversões separados por um traço "-". Para obter a cotação de mais de uma moeda os códigos devem estar separados por vírgula. Abaixo, segue um exemplo de como obter as cotações:

````
#realizando requisição HTTP do tipo GET utilizando a biblioteca "requests" para obter a cotação real do dolar e do euro
cotacao = requests.get("https://economia.awesomeapi.com.br/json/last/USD-BRL,EUR-BRL")

#Conforme mencionado no item configuração, o retorno da solicitação HTTP é um arquivo do tipo json
#Para tratar esse arquivo basta atribuir a váriavel de retorno da solicitação o método json()
#Para esse caso, sobrescrevemos a variável cotacao

#transformando para json
cotacao = cotacao.json()

#imprimindo o arquivo json
print(cotacao)
````

````
#Para obter o valor real do dolar e do euro precisamos encontrar a chave 'bid', que possui o valor da cotação no momento que foi realizado a requisição

#coletando informações da cotação na coluna bid
dolar_real = cotacao['USDBRL']['bid']
euro_real = cotacao['EURBRL']['bid']

print("Cotação do real brasileiro em dolar: ", dolar_real)
print("Cotação real brasileiro em euro: " ,euro_real)
````

#####Retornando Fechamento dos Últimos 7 Dias

Para demonstrar essa utilização básica devemos utilizar o link abaixo, ele retorná o fechamento da moeda especifica nos útimos dias.

Link = https://economia.awesomeapi.com.br/json/daily/:moeda/:numero_dias

No parâmetro número de dias ":numero_dias" que está no link, devemos inserir o número de dias anteriores que queremos obter as informações das moedas específicas. Para o parâmetro ":moedas" devemos seguir da mesma forma que o exemplo anterior.


````
#realizando requisição HTTP do tipo GET para obter a cotação do euro
cotacao = requests.get("https://economia.awesomeapi.com.br/json/daily/EUR-BRL/7")

#transformando para json
cotacao = cotacao.json()

#imprimindo o arquivo json
print(cotacao)
````

````
#Imprimindo os parametros 'bid' que representam a cotação
for i in cotacao:
   print(i['bid'])
````

#### **EXEMPLOS PRÁTICOS**

Suponha que uma pessoa queira comprar um Playstation 5 importado e para economizar ela está fazendo um levantamento do valor do dolar nos últimos 360 dias para saber se agora é um bom momento para realizar a sua compra. Para isso, vamos levantar os dados históricos e realizar uma análise gráfica para verificar a variação do dolar.

````
#Obtendo as informações do dolar no período
cotacao = requests.get("https://economia.awesomeapi.com.br/json/daily/USD-BRL/360")
cotacao = cotacao.json()

#Criando lista para armazenar os valores
valores = []

#Armazenando os valores do dolar em uma lista
for i in cotacao:
   valores.append(i['bid'])

print(valores)
````

````
import matplotlib.pyplot as plt
from datetime import datetime, timedelta


# Data inicial (hoje)
data_atual = datetime.today()

#inserindo range de datas para dos últimos 360 dias
datas = [data_atual - timedelta(days=i) for i in range(360)]

# Convertendo os valores do dolar para números de ponto flutuante
valores = [float(valor) for valor in valores]

# Plotando os valores em relação às datas
plt.figure(figsize=(12, 6))
plt.plot(datas, valores, marker='o', linestyle='-')
plt.xlabel('Data')
plt.ylabel('Valor')
plt.title('Evolução dos Valores em 360 Dias')
plt.grid(True)

# Rotação das datas no eixo x para melhor legibilidade
plt.xticks(rotation=45)

# Exibir o gráfico
plt.tight_layout()
plt.show()
````

Vemos que o valor do dolar vem recuando ao longo dos últimos meses, portanto a pessoa interessada na compra do Playstation 5 pode considerar fazer a compra agora ou aguardar o dolar atingir os valores que tivemos entre julho e setembro.
